```python
#import numpy as np
#import pandas as pd
#import matplotlib.pyplot as plt
import seaborn as sns

#%matplotlib inline
```

```python
# using seaborn load_dataset function to import iris dataset & print the dataset
iris_df = sns.load_dataset('iris')
iris_df
```

```python
# PairPlot method
sns.pairplot(iris_df, hue="species", markers=["o", "s", "D"], diag_kind='kde')
```

```python
# PairGrid method
iris_g = sns.PairGrid(iris_df, hue="species")
iris_g.map_diag(sns.kdeplot)
iris_g.map_offdiag(sns.scatterplot)
iris_g.add_legend()
```
