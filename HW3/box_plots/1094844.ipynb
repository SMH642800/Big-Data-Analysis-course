```python
#import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# load dataset & print the dataset
forestfires_df = pd.read_csv('forestfires.csv')
forestfires_df
```

```python
# use pandas.boxplots
forestfires_df.boxplot(column='FFMC', by='day', rot=45, fontsize=15)
plt.xlabel('Week')
plt.ylabel('FFMC')
plt.title('Pandas Boxplots')
```

```python
# use matplotlib.boxplot & order the week from sunday to saturday
day0 = forestfires_df.loc[forestfires_df['day']=='sun', 'DMC'].values
day1 = forestfires_df.loc[forestfires_df['day']=='mon', 'DMC'].values
day2 = forestfires_df.loc[forestfires_df['day']=='tue', 'DMC'].values
day3 = forestfires_df.loc[forestfires_df['day']=='wed', 'DMC'].values
day4 = forestfires_df.loc[forestfires_df['day']=='thu', 'DMC'].values
day5 = forestfires_df.loc[forestfires_df['day']=='fri', 'DMC'].values
day6 = forestfires_df.loc[forestfires_df['day']=='sat', 'DMC'].values

plt.boxplot([day0, day1, day2, day3, day4, day5, day6], labels=['sun', 'mon', 'tue', 'wed', 'thu', 'fri', 'sat'])
plt.xlabel('Week')
plt.ylabel('DMC')
plt.title('Matplotlib Boxplots')
```

```python
# use seaborn.boxplot & order the week from sunday to saturday
sns.boxplot(data=forestfires_df, x="day", y="DC", order=['sun', 'mon', 'tue', 'wed', 'thu', 'fri', 'sat'])
plt.xlabel('Week')
plt.title('Seaborn Boxplots')
```
